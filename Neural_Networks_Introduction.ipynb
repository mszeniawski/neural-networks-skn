# 1. Trochę teorii...

## 1.1 Biologiczne inspiracje sztucznych sieci neuronowych

Skomplikowane systemy uczące w mózgach zwierząt składają się z połącząnych miedzy sobą grup neuronów. <br><br>

**Stuczne sieci neuronowe** (sieci neuronowe (NN - ang. *neural networks*)) stanowią próbę odwzorowania uczenia nielinowego występującego w biologicznych systemach nerwowych. Sieci neuronowe składają się z określonych grup elementów naśladujących działanie neuronu jako jednostki biologicznej.

![title](obrazki/o1.jpg)

- **Dendryty** - rozgałęziona struktura wielu wejść informacyjnych zbierających sygnały z innych komórek receptorów
- **Akson** - wyprowadza sygnał wyjściowy i przekazuje go do odbiorników informacji
- **Synapsa** - przekazuje sygnał od neuronu do neuronu
- **Korpus komórki nerwowej** - agreguje sygnały wejściowe i wyznacza sygnał wyjściowy

## 1.2 Model sztucznego neuronu

![title](obrazki/o2.jpg)
**Sztuczny neuron**
- Zbiera sygnały (dane) wejściowe ($ x_{i} $) z poprzedzających go neuronów
- Sumuje otrzymane sygnały z odpowiednimi wagami ($ w_{i} $) i tworzy z nich jedną wartość (np. za pomocą funkcji $ \sum $)
- Otrzymany wynik jest wyjściem dla funkcji aktywacji ($ \varphi  $)
- Funkcja aktywacji zwraca sygnał wyjściowy (dane wyjściowe) ($ y $)
- Sygnał wyjściowy przekazywany jest do następnych neuronów

Neuron: <br>
- stanowi jednostkę przetwarzającą przejmującą $n$ sygnałów wejściowych
- sygnał przetwarza z wykorzystaniem wag $ w_{i}, i = 1, ... n $
- wyznacza wartość sygnału wyjściowego $y$ według wzoru: <br>

## $y = \varphi (u)$

## $u = w_{0} + \sum_{i=1}^{n}w_{i}x_{i} $

Jednymi z pierwszych, którzy zaproponowali model sztucznego neuronu byli Warren McCulloch i Walter Pitts. W przypadku modelu $ \varphi, $ będące funkcją aktywacji ma postać funkcji skokowej Heaviside’a: <br>

## 1.3 Model perceptronu

Perceptron stanowi jedną z najprostrzysz architektur SSN (sztucznych sieci neuronowych). Perceptron składa się po prostu z jednej warstwy jednostek **LTU** (jednostka liniowa z progiem; neuron; ang. *linear treshold unit*). gdzie każdy neuron jest połączony ze wszystkimi wejściami. Połączenie te są często reprezentowane przez specyficzne neurony przechodnie, zwane **neuronami wejściowymi** (ang. *input neuron*) <br><br>

Na rysunku widzimy perceptron z dwoma wejściami i trzema wyjściami. Może on jednocześnie klasyfikować próbki do trzech różnych klas binarnych, czyli stanowi on klasyfikator wielowyjściowy.
<br>

### 1.3.1 Trenowanie perceptronu

**Reguła Hebba** - * "Komórki pobudzające się wzajemnie wiążą się ze sobą" * Zgodnie z regułą waga połączeń pomiędzy dwoma neuronami wzrasta za każdym razem, gdy mają ten sam stan na wyjściu. <br> <br>
Perceptrony są uczone za pomocą odmiany tej reguły, w której brany jest pod uwagę błąd popełniany przez sieć; nie są wzmacniane połączenia prowadzące do nieprawidłowych wyników. Dokładniej mówiąć, perceptron przetwarza naraz jeden przykład uczący i wylicza dla niego prognozy. Na każdy neuron wyjściowy odpowiedzialny za nieprawidłowy wynik następuje zwiększenie wag połączeń ze wszystkimi wejściami przyczyniającymi się do właściwej prognozy

## $ w_{i,j}^{\ \ \ (nastepny \ krok)}= w_{i,j} + \eta (y_{j} - \hat{y_{j}} )x_{i} $

-  $w_{i,j} $  - waga połączenia pomiędzy $i$-tym neuronem wejściowym i $j$-tym neuronem wyjściowym
- $ x_{i} $ - $i$-ta wartość wejściowa bieżącego przykładu uczącego
- $ \hat{y_{j}} $ - wynik $j$-tego neuronu wyjściowego dla bieżącego przykładu uczącego
- $ y_{j} $ - docelowy wynik $j$-tego neuronu wyjściowego dla bieżącego przykładu uczącego
- $ \eta $ - współczynnik uczenia

### 1.3.2 Przykład

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

iris = load_iris()
X = iris.data[:, (2, 3)]  # długość płatka, szerokość płatka
y = (iris.target == 0).astype(np.int)

per_clf = Perceptron(random_state=42)
per_clf.fit(X, y)

y_pred = per_clf.predict([[2, 0.5]])
```

## 1.4 Perceptron wielowarstwowy i propagacja wsteczna

Perceptron **MLP** (percepton wielowarstwowy; ang. *multi-layer perceptron*) składa się z jednej **warstwy wejściowej** (przechodniej), co najmniej jednej warstwy jednostek LTU (tzw. **warstwy ukryte**) i ostatniej warstwy jednostek LTU (**warstwa wyjściowa**)

<img src="obrazki/o4.jpg" width="500px" height="500px">

Oprócz warstwy wyjściowej każda warstwa zawiera neuron obciążający oraz jest w pełni połączona z następną warstwą. Gdy sieć SSN zawiera przynajmniej dwie warstwy ukryte, nosi nazwę **głębokiej sieci neuronowej** (GSN)

### 1.4.1 Trenowanie perceptronów wielowarstwowych

**Algorytm wstecznej propagacji** (ang. *backpropagation*)

1. Algorytm pobiera każdy przykład uczący do sieci i oblicza wynik wszystkich neuronów w każdej następnej warstwie (jest to przebieg do przodu, taki sam, jak podczas wyliczania prognoz). 
2. Następnie zostaje zmierzony błąd na wyjściu sieci (tj. różnicza pomiędzy oczekiwanym a uzyskanym wynikiem), a także wpływ każdego neuronu w ostatniej wartwie ukrytej na błędny wynik każdego neuronu wyjściowego. 
3. W dalszej kolejności zostaje zmierzony wpływ neuronów ze wcześniejszej warstwy ukrytej - i tak dalej, aż do osiągnięcia warstwy wejściowej. 

Taki odwrotny przebieg skutecznie mierzy gradient błędu we wszystkich wagach połączeń poprzez propagację tego gradientu w kierunku początku sieci.<br>
Ostatnim etapem algorytmu propagacji wstecznej jest mechanizm gradientu prostego wobec wszystkich wag połączeń w seci za pomocą zmierzonego wcześniej gradientu błędu. <br>
Żeby ten algorytm działał prawidłowo, twórcy musieli wprowadzić kluczową zmianę w architekturze MLP; zastąpili funkcję skokową funkcją logistyczną:

# $ \sigma (z) = \frac{1}{1 + e^{-z}} $

Było to niezbędne, ponieważ funkcja skokowa zawiera jedynie płaskie segmenty, dlatego nie pozwala korzystać z gradientu (algorytm gradientu nie może się poruszać po płaskim przebiegu funkcji), natomiast funkcja logistyczna ma w każdym punkcie zdefiniowaną pochodną niezerową, dzięki czemu algorytm gradientu prostego może w każdym etapie uzyskiwać lepsze wyniki.

Inne wykorzystywane funkcje aktywacji:

- **funkcja tangensa hiperbolicznego**: $ tanh(z) = 2\sigma (2z)-1 $
- **funckja ReLU**: $ ReLU(z) = max(0,z) $

## 3.2 Model sieci neuronowej

### 3.2.1 Przygotowanie danych

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

# ustawienie ziarna generatora losowego
np.random.seed(7)
```

```python
X = diabets_data.iloc[:,0:8]
Y = diabets_data.iloc[:,8]
```

```python
X.head()
```

```python
Y.head()
```

### 3.2.2 Konstrukcja i trenowanie sieci

```python
# konstruowanie modelu
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# kompilacja modelu
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#trenowanie
#batch_size -> https://stats.stackexchange.com/questions/153531/what-is-batch-size-in-neural-network
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
#ocena modelu
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

### 3.2.3 Wizualizacji procesu trenowania

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(X, Y, validation_split=0.33, epochs=150, batch_size=10, verbose=0)
```

```python
history_frame = pd.DataFrame(history.history)
history_frame["epoch"] = history.epoch
```

```python
history_frame.head()
```

```python
import matplotlib.pyplot as plt

print(history.history.keys())

# accuracy na zbiorze testowym i walidacyjnym
plt.figure(figsize=(10,5))
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# loss na zbiorze testowym i walidacyjnym
plt.figure(figsize=(10,5))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

## 4.2 Model sieci neuronowej

### 4.2.1 Przygotowanie danych

```python
X = train.iloc[:,0:9]
Y = train.iloc[:,9]
```

```python
X.head()
```

```python
Y.head()
```

### 4.2.2 Konstrukcja i trenowanie sieci

```python
import keras.optimizers as opt
from sklearn.preprocessing import StandardScaler
from keras import backend
 
def rmse(y_true, y_pred):
    return backend.sqrt(backend.mean(backend.square(y_pred - y_true), axis=-1))

def baseline_model():
    
    model = Sequential()
    model.add(Dense(300, input_dim=9, activation='relu'))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(25, activation='relu'))
    model.add(Dense(1))

    model.compile(loss='mean_squared_logarithmic_error', optimizer=opt.Adamax(lr=0.02, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0), metrics=[rmse])
    return model
```

```python
base_model = baseline_model()
```

```python
base_model.fit(X, Y, epochs=250, batch_size=30)
```

```python
base_model.evaluate(X, Y)
```

### 4.2.3 Standaryzacja

```python
import warnings
warnings.filterwarnings('ignore')

mat_X = np.matrix(X)
mat_y = np.matrix(Y)

X = StandardScaler().fit_transform(mat_X)
Y = StandardScaler().fit_transform(mat_y)
Y = np.reshape(Y, (1460,1))
```

```python
X
```

```python
Y
```

```python
std_model = baseline_model()
```

```python
std_model.fit(X, Y, epochs=250, batch_size=30)
```

```python
std_model.evaluate(X, Y)
```

```python
def new_model():
    
    model = Sequential()
    model.add(Dense(300, input_dim=9, activation='relu'))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(25, activation='relu'))
    model.add(Dense(1))
    
    model.compile(loss='mean_squared_error', optimizer="adam", metrics=[rmse])
    return model
```

```python
nw_model = new_model()
```

```python
nw_model.fit(X, Y, epochs=250, batch_size=30)
```

```python
nw_model.evaluate(X, Y)
```
